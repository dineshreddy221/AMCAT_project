## Hypothesis Testing

<B>Times of India article dated Jan 18, 2019 states that “After doing your Computer Science Engineering if you take up jobs as a Programming Analyst, Software Engineer, Hardware Engineer and Associate Engineer you can earn up to 2.5-3 lakhs as a fresh graduate.” Test this claim with the data given to you.</B>

$ Step - 1: $
<li> Alternate Hypothesis (Bold claim):</li>

$$ H_1:\mu < 300000$$

<li> Null Hypothesis(Status Quo)</li>

$$H_0:\mu \geq 300000$$

$ Step - 2: $$

<li> Sample of size</li>
$$ n= 35 $$

<li>Compute sample mean</li>
$$ \bar{x} = 297000.0 $$

$ Step - 3: $

<li> Compute Test Statistics</li>

$$ t = \frac{\bar{x} - \mu}{S\ /\sqrt[2]{n}} $$

$ step - 4: $

<li> Decide $ \alpha $ or significance level</li>
$$ \alpha = 0.05 $$

$ Step - 5.1: $

<li>one tailed test:</li>
$$ Left\ tailed\ test\ $$

$ Step - 5.2: $

<li> compute p-value</li>

$$ p-value = -0.0 $$


```python
from scipy.stats import norm
```

```python
# score for Sampling distribution

def z_score(sample_size, sample_mean, pop_mean, pop_std):
    numerator=sample_mean - pop_mean
    denominator = pop_std /sample_size**0.5
    return numerator / denominator
```

```python
# calculating sample mean using computer science & engineering graduates salary
Sample=df['Salary'][df['Specialization']=='computer science & engineering'].iloc[0:35]
```

```python
sample_1=list(Sample)
```

```python
# calculating sample mean
sample_mean=sum(Sample)/len(Sample)
sample_mean
```

```python
# calculating sample standard deviation
sample_size=35
for i in range(len(sample_1)):
    sample_std=((sample_1[i]-sample_mean)**2)**0.5/(sample_size-1)          
print(sample_std)
```

```python
sample_size = 35
sample_mean = 297000
pop_mean = 300000
sample_std =2852.94 
```

```python
# defining the z_score
z=z_score(sample_size, sample_mean, pop_mean, sample_std)
print(z)
```

```python
confidence_level = 0.95

alpha = 1 - confidence_level

z_critical = norm.ppf(1 - alpha/2)

print(z_critical)
```

```python
x_min = 200000
x_max = 500000


# Defining the sampling distribution mean and sampling distribution std
mean = pop_mean
std = sample_std / sample_size**0.5


# Ploting the graph and setting the x limits
x = np.linspace(x_min, x_max, 100)
y = norm.pdf(x, mean, std)
plt.xlim(x_min, x_max)
plt.plot(x, y)


# Computing the left critical values (one tailed Test)
z_critical_left = pop_mean + (-z_critical * std)


# Shading the left rejection region
x1 = np.linspace(x_min, z_critical_left, 100)
y1 = norm.pdf(x1, mean, std)
plt.fill_between(x1, y1, color='orange')


# Ploting the sample mean and concluding the results 
plt.scatter(sample_mean, 0)
plt.annotate("x_bar", (sample_mean, 0.00001))
```

```python
# conclusion using z_test

if (np.abs(z) > z_critical):
    print("Reject Null Hypothsis")
else:
    print("Fail to reject Null Hypothesis")
```

```python
# conclusion using p_test

p_value = z*(1.0 - norm.cdf(np.abs(z)))

print("p_value =", p_value)

if (p_value < alpha):
    print("Reject Null Hypothesis")
else:
    print("Fail to reject Null Hypothesis")
```
